### DataFrame

Any Tabular data in python which is loaded with pandas is known as DataFrame.
Any 2D structure of data consisting of rows and columns is known as DataFrame.
- Single row/column is called Series

 **Creating DataFrame**

```python
# Using Lists

student_data = [
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,50,2]
]

pd.DataFrame(student_data,columns=['iq','marks','package'])   
```

```python
# Using Dict

student_dict = {
    'name':['Example1','Example2','Example3','Example4','Example5','Example6'],
    'iq':[100,90,120,80,0,0],
    'marks':[80,70,100,50,0,0],
    'package':[10,7,14,2,0,0]
}
student = pd.DataFrame(student_dict)
student.set_index('name',inplace = True)
student
```

### Selecting rows from a DataFrame

- **iloc** - searches using index positions
- **loc** - searches using index lab

Diff b/w iloc and loc is that loc includes the last index for printingls 

```python
# single row
movies.iloc[101]
```

```python
# multiple rows

movies.iloc[-9::2]
```

```python
# Fancy Indexing

movies.iloc[[1,234,789]]
```

```python
# loc
# Single row
student.loc['Example2']
```

```python
# Multiple row
student.loc['Example1':'Example5':2] # includes the last number also
```

```python
# Fancy indexing

student.loc[['Example1','Example3','Example4']]
```

```python
def check_rec(Team_1,Team_2):
    rec = (ipl['Team1']== Team_1) & (ipl['Team2']==Team_2)
    return ipl[rec]
```

```python
check_rec('Kolkata Knight Riders','Royal Challengers Bangalore')
```
